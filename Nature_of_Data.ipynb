Analysis of Categorical Variables
1. Job
2. Marital Status
3. Housing Loan
4. outcome of the previous marketing campaign (poutcome)
6. day of week
7. education

```python
def plot_distribution_by_category(data_set, category_column, title):
    # Create separate DataFrames for success and failure
    success_data = data_set[data_set['y'] == 'yes']
    failure_data = data_set[data_set['y'] == 'no']

    # Get unique categories
    categories = data_set[category_column].value_counts().index

    # Initialize subplots
    fig, ax = plt.subplots(figsize=(12, 6))

    # Width of the bars
    bar_width = 0.35

    # Index for the x-axis positions
    indices = range(len(categories))

    # Plot bars for success and failure side by side
    success_bars = ax.bar(indices, [success_data[success_data[category_column] == category].shape[0] for category in categories], 
                          width=bar_width, color='skyblue', alpha=0.7, label='Success')

    failure_bars = ax.bar([i + bar_width for i in indices], [failure_data[failure_data[category_column] == category].shape[0] for category in categories], 
                          width=bar_width, color='hotpink', alpha=0.7, label='Failure')

    # Set plot properties
    ax.set_title(f'{title} Distribution by Success/Failure')
    ax.set_xlabel(category_column.capitalize())
    ax.set_ylabel('Count')
    ax.set_xticks([i + bar_width/2 for i in indices])
    ax.set_xticklabels(categories, rotation=45, ha='right')

    # Add legend
    ax.legend(title='Outcome', loc='upper right', labels=['Success', 'Failure'])

    plt.show()

column_names = list(bank_data_set)
column_names.remove('y')

title_names = ['Age','Job','Martial','Education','Default','Housing','Loan','Contact','Month','Day Of Week', 'Duration','Campaign','Previous Days','Previous','Previous Outcome','Employment Variation Rate','Consumer Price Index', 'Consumer Confidence Index','Euribor 3 Months Rate','Number of Employees']

for column_name, title_name in zip(column_names, title_names):
    plot_distribution_by_category(bank_data_set, column_name, title_name)
```

- More customers working as admin than any other profession
- For each job failure is way higher than success
- Majority of the customers are married. Followed by Single, divorced and unknown
- Majority of the customers have a housing loan
- Most of the customers are new customers who have not been contacted earlier
- Majority of customers who had a successful outcome from the prev. campaign, did subscribe for a term deposit
- Day of Week seems to be irrelevant
- Most people in the data set have a university degree

Analysis of Numerical Variables
1. Age
2. Duration (Of the last contact, in seconds)
3. euribor3m (basic rate of interest used in lending between banks on the European Union interbank market)

```python
def plot_histogram_and_means(data_set, column, title, xlabel, unit):
    # Create separate DataFrames for success and failure
    success_data = data_set[data_set['y'] == 'yes']
    failure_data = data_set[data_set['y'] == 'no']

    # Plot the histograms for success and failure
    plt.figure(figsize=(12, 8))

    plt.hist(success_data[column], bins=20, alpha=0.8, label='Success', color='blue', edgecolor='black')
    plt.hist(failure_data[column], bins=20, alpha=0.4, label='Failure', color='hotpink', edgecolor='black')

    plt.title(f'{title} Distribution by Success/Failure')
    plt.xlabel(xlabel)
    plt.ylabel('Count')

    # Add legend
    plt.legend()

    plt.show()

    # Calculate the mean values
    mean_success = success_data[column].mean()
    mean_failure = failure_data[column].mean()

    print(f'The mean {xlabel.lower()} for people who subscribed is: {mean_success:.2f} {unit}')
    print(f'The mean {xlabel.lower()} for people who did not subscribe is: {mean_failure:.2f} {unit}')

# AGE
plot_histogram_and_means(bank_data_set, 'age', 'Age', 'Age', 'years')

# DURATION
plot_histogram_and_means(bank_data_set, 'duration', 'Duration', 'Duration', 'sec')

# EURIBOR3M
plot_histogram_and_means(bank_data_set, 'euribor3m', 'euribor3m rate', 'euribor3m rate', '%')
```

- Age isn’t necessarily a good indicator for which customer will subscribe and which customer will not
- Last contact duration of a customer can be useful for predicting the target variable
- Difference in euribor3m median for both the classes
- euribor3m could be very useful for our case study
